```python
import random
t0 = pd.to_datetime("1980-01-01")

states = [ "AA", "BB", "CC"]
statescols = {"AA":"r", "BB":"g", "CC":"b" }

data = [ ]
for childn in range(5000):
    n = random.randint(1, 5)
    ta = t0 + pd.Timedelta(days=random.randint(4000, 9000))

    for i in range(n):
        tb = ta + pd.Timedelta(days=random.randint(10, 1000))
        st = random.choice(states)
        data.append((childn, ta, tb, st))
        ta = tb
    
```
